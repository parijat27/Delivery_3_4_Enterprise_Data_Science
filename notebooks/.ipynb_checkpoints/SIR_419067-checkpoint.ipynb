```python
import pandas as pd
import numpy as np
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

from scipy import optimize
from scipy import integrate

import plotly.graph_objects as go
import plotly
pd.set_option('display.max_rows',1000)

from datetime import datetime
import seaborn as sns

sns.set(style="darkgrid")

mpl.rcParams['figure.figsize'] = (16, 9)
pd.set_option('display.max_rows', 1000)
```

```python
df_analyse = pd.read_csv('../data/processed/COVID_small_flat_table.csv', sep = ',', parse_dates=[0])
df_analyse.sort_values('date', ascending=True).head()
```

```python
# set some basic parameters
# beta/gamma is denoted as  'basic reproduction number'

N0=1000000 #max susceptible population
beta=0.4   # infection spread dynamics
gamma=0.1  # recovery rate


# condition I0+S0+R0=N0
I0=df_analyse.Germany[35]
S0=N0-I0
R0=0
```

```python
def SIR_model(SIR,beta,gamma):
    ''' Simple SIR model
        S: susceptible population
        I: infected people
        R: recovered people
        beta: 
        
        overall condition is that the sum of changes (differnces) sum up to 0
        dS+dI+dR=0
        S+I+R= N (constant size of population)
    
    '''
    
    S,I,R=SIR
    dS_dt=-beta*S*I/N0          #S*I is the 
    dI_dt=beta*S*I/N0-gamma*I
    dR_dt=gamma*I
    return([dS_dt,dI_dt,dR_dt])
```

# Simulative approach to calculate SIR curves


```python
SIR=np.array([S0,I0,R0])
propagation_rates=pd.DataFrame(columns={'susceptible':S0,
                                        'infected':I0,
                                        'recoverd':R0})



for each_t in np.arange(100):
   
    new_delta_vec=SIR_model(SIR,beta,gamma)
   
    SIR=SIR+new_delta_vec
    
    propagation_rates=propagation_rates.append({'susceptible':SIR[0],
                                                'infected':SIR[1],
                                                'recovered':SIR[2]}, ignore_index=True)
```

```python
fig, ax1 = plt.subplots(1, 1)

ax1.plot(propagation_rates.index,propagation_rates.infected,label='infected',color='k')
ax1.plot(propagation_rates.index,propagation_rates.recovered,label='recovered')
ax1.plot(propagation_rates.index,propagation_rates.susceptible,label='susceptible')

ax1.set_ylim(10, 1000000)
ax1.set_yscale('linear')
ax1.set_title('Scenario SIR simulations  (demonstration purposes only)',size=16)
ax1.set_xlabel('time in days',size=16)
ax1.legend(loc='best',
           prop={'size': 16});
```

```python
def SIR_model_t(SIR,t,beta,gamma):
    ''' Simple SIR model
        S: susceptible population
        t: time step, mandatory for integral.odeint
        I: infected people
        R: recovered people
        beta: 
        
        overall condition is that the sum of changes (differnces) sum up to 0
        dS+dI+dR=0
        S+I+R= N (constant size of population)
    
    '''
    
    S,I,R=SIR
    dS_dt=-beta*S*I/N0          #S*I is the 
    dI_dt=beta*S*I/N0-gamma*I
    dR_dt=gamma*I
    return dS_dt,dI_dt,dR_dt
```

```python
def fit_odeint(x, beta, gamma):
    '''
    helper function for the integration
    '''
    return integrate.odeint(SIR_model_t, (S0, I0, R0), t, args=(beta, gamma))[:,1] # we only would like to get dI
```

# Dynamic beta in SIR (infection rate)


```python
country_list = ['Italy',
               'US',
               'Spain',
               'Germany',
               'Korea, South',]

i=1
final_data = []
ydata_total_total = []
#final_data = pd.DataFrame(fitted)
temp = []
for each in country_list:
    country = df_analyse.columns[i]
    ydata_total = np.array(df_analyse[country])[40:200]
    t=np.arange(len(ydata_total))
    
    popt, pcov = optimize.curve_fit(fit_odeint, t, ydata_total)
    perr = np.sqrt(np.diag(pcov))
    
    print('standard deviation errors : ',str(perr), ' start infect:',ydata_total[0])
    print("Optimal parameters: beta =", popt[0], " and gamma = ", popt[1])
    
   

    fitted=fit_odeint(t, *popt)
    # option 1: ydata is cummulative, thus we need cumulative fitted line
    #plt.semilogy(t, ydata_total, 'o')
    #plt.semilogy(t, fitted)

    # option2: ydata is cummulative, however we would like to have the day to day difference
    plt.semilogy(t,np.ediff1d(ydata_total, to_begin=ydata_total[1]-ydata_total[0]), 'o')
    plt.semilogy(t, fitted)
    
    #temp = pd.DataFrame(fitted)
    final_data.append(fitted)
    #final_data.loc[:, each] = fitted
    
    final_data_df = pd.DataFrame(final_data)
    final_data_df = final_data_df.T
    
    ydata_total_total.append(ydata_total)
    ydataa_total_df = pd.DataFrame(ydata_total_total)
    ydataa_total_df = ydataa_total_df.T
    
    i=i+1
    
final_data_df.columns =country_list
ydataa_total_df.columns = country_list    
    
    
plt.title("Fit of SIR model for Germany cases")
plt.ylabel("Population infected")
plt.xlabel("Days")
plt.show()
#print("Optimal parameters: beta =", popt[0], " and gamma = ", popt[1])
#print("Basic Reproduction Number R0 " , popt[0]/ popt[1])
#print("This ratio is derived as the expected number of new infections (these new infections are sometimes called secondary infections from a single infection in a population where all subjects are susceptible. @wiki")
```

```python
import dash
from dash import dcc
from dash import html
from dash.dependencies import Input, Output

app1 = dash.Dash()
app1.layout = html.Div([
    
    #html.Label('Select Country to display fitted data'),
    
    dcc.Markdown('''
    ## Delivery 4: Select Country to display fitted data
    '''),
    dcc.Markdown('''
    ### The following chart displays the fitted values of the COVID-19 infection data for SIR dynamic simulation
    '''),
    
    dcc.Dropdown(
        id = 'country_drop',
        options = [
            {'label': 'Italy', 'value' : 'Italy'},
            {'label': 'US', 'value' : 'US'},
            {'label': 'Spain', 'value' : 'Spain'},
            {'label': 'Germany', 'value' : 'Germany'},
            {'label': 'Korea, South', 'value' : 'Korea, South'}
        ],
        value = ['Spain', 'Germany','Italy'],
        multi = True

    ),
    
    dcc.Graph(figure = fig1,id='window1')
])

@app1.callback(
    Output('window1', 'figure'),
    [Input('country_drop', 'value')])


def update_figure(country_list):
    
    traces = []

    for each in country_list:

        traces.append(dict(x=np.arange(len(ydata_total)),
                                y=final_data_df[each],
                                mode='markers+lines',
                                line_width = 2, 
                                marker_size = 1,
                                opacity=0.9,
                                name=each + '_fitted_cases'
                        )
                )
        
        traces.append(dict(x=np.arange(len(ydata_total)),
                                y=df_analyse[each],
                                mode='lines',
                                opacity=0.9,
                                line_width = 6, 
                                name=each + '_total_cases'
                        )
                )

        traces.append(dict(x=np.arange(len(ydata_total)),
                                y=np.ediff1d(ydataa_total_df[each], to_begin=ydataa_total_df[each][1]-ydataa_total_df[each][0]),
                                mode='markers',
                                opacity=0.9,
                                name=each+'_difference_cases'
                        )
                )
        
        
        
        
    return {
            'data': traces,
            'layout': dict (
                width=1400,
                height=600,

                xaxis={'title':'Days',
                        'tickangle':-45,
                        'nticks':20,
                        'tickfont':dict(size=14,color="#7f7f7f"),
                      },
                
                yaxis={'type':"log",'title':'Population Infected'}

        )
    }


app1.run_server(debug=True, use_reloader=False, port = 8006)
```
