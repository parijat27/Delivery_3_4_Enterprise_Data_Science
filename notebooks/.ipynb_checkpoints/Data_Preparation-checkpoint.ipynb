```python
country_list = ['Italy',
               'US',
               'Spain',
               'Germany',
               'Korea, South',]
```

```python
# Import population data
wp = pd.read_csv('../data/raw/COVID-19-master/data_tables/world_pop_by_country.csv')
wp.head()
```

```python
# Take summation

for each in country_list:
    df_plot[each] = np.array(pd_raw[pd_raw['Country/Region'] == each].iloc[:,4::].sum(axis=0))
    var = np.array(wp.loc[wp["Country Name"]==each,"2018"])
    var = var.astype(int)

    print(var)
    df_plot[each]=df_plot[each]/var[0]*100
```

```python
%matplotlib inline
plt.figure
df_plot.set_index('date').plot()
plt.ylim = (10,30000)
```

## Relational Data model

```python
pd_raw.head()
```

```python
pd_data_base = pd_raw.rename(columns={'Country/Region' : 'country',
                            'Province/State' : 'state'})
pd_data_base['state']=pd_data_base['state'].fillna('no')  #ensure a sring, important for indexing
```

```python
pd_data_base = pd_data_base.drop(['Lat', 'Long'],axis=1)
pd_data_base.head()
```

```python
pd_relational_model = pd_data_base.set_index(['state', 'country']) \
                                            .T                     \
                                            .stack(level = [0,1])  \
                                            .reset_index()         \
                                            .rename(columns={'level_0' : 'date' ,
                                                            0 : 'confirmed'},
                                                   )
pd_relational_model.head()
```

```python
pd_relational_model.dtypes
```

```python
pd_relational_model['date']=pd_relational_model.date.astype('datetime64[ns]')
```

```python
pd_relational_model.dtypes
```

```python
pd_relational_model.confirmed=pd_relational_model.confirmed.astype(int)

pd_relational_model.to_csv('../data/processed/COVID_relational_confirmed.csv',sep=';',index=False)
```
